**series**
a series is a one dimentional labelled array that can hold any datatype 
like single coulmn in a spreadseet

```python
import pandas as pd
data=[100.1,200.2,300,400]
series = pd.Series(data)
print(series)
```

datatype mentioned below is the data type of the elements

we can change the index number also using the index[] here we can use the another list,tuple,string,another series

```python
series=pd.Series(data,index=["a","b","c","d"])
print(series)
```

Creating a DataFrame by passing a NumPy array with a datetime index using date_range() and labeled columns:


```python
import numpy as np 
s=pd.date_range(20260611,periods=6)
s

df=pd.DataFrame(np.random.randn(6,4),index=s,columns=list("ABCD"))
df
```

if we didn't mention any index it will take the index by defaukt in form of 1,2,3 
we can pass the dictionary with the key value pairs so key is considered as the column column label and values will be conoisdered as column value

```python
print(df.head(2))
print(df.tail())
```

Display the DataFrame.index or DataFrame.columns:

```python
print(df.index)
print(df.columns)
```
